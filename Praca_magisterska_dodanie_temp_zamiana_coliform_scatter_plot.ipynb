```python
# Vi to wartość parametru- próbki wody

# wartości referencyjne 
Videal = {
  'D.O. (mg/l)': 14.6,                   
  'PH': 7.0,                             
  'CONDUCTIVITY (µmhos/cm)': 0,      
  'B.O.D. (mg/l)': 0,                     
  'NITRATENAN N+ NITRITENANN (mg/l)': 0, 
  'FECAL COLIFORM (MPN/100ml)': 0,        
  'TOTAL COLIFORM (MPN/100ml)Mean': 0  
}

# wartość oczekiwana dla każdego z parametrów czystej wody 
Si = {
    'D.O. (mg/l)': 10,                   
    'PH': 8.5,                             
    'CONDUCTIVITY (µmhos/cm)': 1000,        
    'B.O.D. (mg/l)': 45,                     
    'NITRATENAN N+ NITRITENANN (mg/l)': 5,  
    'FECAL COLIFORM (MPN/100ml)': 100,        
    'TOTAL COLIFORM (MPN/100ml)Mean': 1000 
}



# waga każdego z parametrów
def calculate_wi(S_i,valid_Si):
    # stała proporcjonalności
    K = 1/sum(valid_Si.values())
    return (K/S_i) 

# skala oszacowania jakości
def calculate_qi(Vi, param):
    return 100 * (Vi - Videal.get(param)) / (Si.get(param) - Videal.get(param))

# indeks jakości wody
def calculate_wqi(row):
    sum_qi_wi = 0
    sum_wi = 0



    valid_Si = Si.copy()
    for param in ['D.O. (mg/l)', 'PH', 'CONDUCTIVITY (µmhos/cm)', 'B.O.D. (mg/l)', 'NITRATENAN N+ NITRITENANN (mg/l)', 
                  'FECAL COLIFORM (MPN/100ml)', 'TOTAL COLIFORM (MPN/100ml)Mean']:   
        if pd.isna(row[param]):
            valid_Si[param] = 0
   
    for param in ['D.O. (mg/l)', 'PH', 'CONDUCTIVITY (µmhos/cm)', 'B.O.D. (mg/l)', 'NITRATENAN N+ NITRITENANN (mg/l)', 
                  'FECAL COLIFORM (MPN/100ml)', 'TOTAL COLIFORM (MPN/100ml)Mean']:                                  
        if not pd.isna(row[param]):
            qi = calculate_qi(row[param], param)
            wi = calculate_wi(Si[param], valid_Si)
            sum_qi_wi += (qi * wi)
            sum_wi += wi
    return round(sum_qi_wi / sum_wi,2)
    
print(len(df_filtered))
df_filtered.shape

for idx, row in df_filtered.iterrows():
    if idx in df_filtered.index:
        df_filtered.loc[idx, 'WQI'] = calculate_wqi(row)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import optuna
from sklearn.model_selection import RepeatedStratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, accuracy_score, confusion_matrix

def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 50, 500)  #zmienność liczby drzew
    max_depth = trial.suggest_int('max_depth', 5, 50)  #głębokość drzewa
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)  #minimalna liczba próbek do podziału
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)  #minimalna liczba próbek na liściu
    max_features = trial.suggest_float('max_features', 0.1, 1.0)  #procent cech używanych do podziału
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])  #użycie bootstrappingu
    criterion = trial.suggest_categorical('criterion', ['gini', 'entropy'])  #kryterium podziału

    X = df_imputated.iloc[:, 3:10]
    Y = df_imputated.iloc[:, 13]

    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        bootstrap=bootstrap,
        criterion=criterion,
        random_state=42
        
    )

    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=99)

    score = cross_val_score(model, X, Y, scoring='accuracy', cv=cv, n_jobs=-1).mean()
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)  

print("Najlepsze parametry:", study.best_params)
print("Najlepszy wynik (accuracy):", study.best_value)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, accuracy_score, confusion_matrix
from sklearn.svm import SVC 
from sklearn.preprocessing import StandardScaler

def objective(trial):
    C = trial.suggest_loguniform('C', 1e-5, 1e5)  #regularyzacja
    kernel = trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid'])  #typ jądra
    degree = trial.suggest_int('degree', 2, 5)  #stopień dla jądra 'poly'
    gamma = trial.suggest_categorical('gamma', ['scale', 'auto'])  #parametr gamma
    coef0 = trial.suggest_float('coef0', 0.0, 1.0)  #stała jądra 'poly' i 'sigmoid'
    shrinking = trial.suggest_categorical('shrinking', [True, False])  #algorytm przyspieszania
    probability = trial.suggest_categorical('probability', [True, False])  #obliczanie prawdopodobieństw
    tol = trial.suggest_float('tol', 1e-5, 1e-1)  #tolerancja na błędy
    max_iter = trial.suggest_int('max_iter', -1, 10000)  #maksymalna liczba iteracji

    X = df_imputated.iloc[:, 3:10]
    Y = df_imputated.iloc[:, 13]

    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    model = SVC(
        C=C,
        kernel=kernel,
        degree=degree,
        gamma=gamma,
        coef0=coef0,
        shrinking=shrinking,
        probability=probability,
        tol=tol,
        max_iter=max_iter,
        random_state=42
    )

    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=99)

    score = cross_val_score(model, X, Y, scoring='accuracy', cv=cv, n_jobs=-1).mean()
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50) 

print("Najlepsze parametry:", study.best_params)
print("Najlepszy wynik (accuracy):", study.best_value)
```

```python
from sklearn.neural_network import MLPClassifier

def objective(trial):
    hidden_layer_sizes = trial.suggest_categorical('hidden_layer_sizes', [(50,), (100,), (50, 50), (100, 50)])
    activation = trial.suggest_categorical('activation', ['relu', 'tanh', 'logistic'])  #funkcja aktywacji
    solver = trial.suggest_categorical('solver', ['adam', 'sgd', 'lbfgs'])  #algorytm optymalizacji
    alpha = trial.suggest_loguniform('alpha', 1e-5, 1e-1)  #regularyzacja
    batch_size = trial.suggest_categorical('batch_size', ['auto', 32, 64, 128])  #rozmiar wsadu
    learning_rate = trial.suggest_categorical('learning_rate', ['constant', 'invscaling', 'adaptive'])  #typ tempa uczenia
    learning_rate_init = trial.suggest_loguniform('learning_rate_init', 1e-5, 1e-1)  #początkowa wartość tempa uczenia
    power_t = trial.suggest_float('power_t', 0.1, 0.9)  #parametr dla 'invscaling'
    max_iter = trial.suggest_int('max_iter', 100, 10000)  #maksymalna liczba iteracji
    shuffle = trial.suggest_categorical('shuffle', [True, False])  #losowanie danych
    tol = trial.suggest_float('tol', 1e-5, 1e-1)  #tolerancja na błąd
    momentum = trial.suggest_float('momentum', 0.5, 0.99)  #parametr momentum dla 'sgd'
    nesterovs_momentum = trial.suggest_categorical('nesterovs_momentum', [True, False])  #użycie Nesterova
    early_stopping = trial.suggest_categorical('early_stopping', [True, False])  #wczesne zatrzymanie
    validation_fraction = trial.suggest_float('validation_fraction', 0.1, 0.3)  #udział danych do walidacji przy wczesnym zatrzymaniu
    beta_1 = trial.suggest_float('beta_1', 0.5, 0.9)  #parametr beta dla 'adam'
    beta_2 = trial.suggest_float('beta_2', 0.9, 0.999)  #parametr beta dla 'adam'
    epsilon = trial.suggest_loguniform('epsilon', 1e-8, 1e-4)  #parametr epsilon dla 'adam'
    n_iter_no_change = trial.suggest_int('n_iter_no_change', 5, 20)  #liczba iteracji bez zmiany do zatrzymania
    max_fun = trial.suggest_int('max_fun', 10000, 15000)  #maksymalna liczba wywołań funkcji w 'lbfgs'


    X = df_imputated.iloc[:, 3:10]
    Y = df_imputated.iloc[:, 13]

    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    model = MLPClassifier(
        hidden_layer_sizes=hidden_layer_sizes,
        activation=activation,
        solver=solver,
        alpha=alpha,
        batch_size=batch_size,
        learning_rate=learning_rate,
        learning_rate_init=learning_rate_init,
        power_t=power_t,
        max_iter=max_iter,
        shuffle=shuffle,
        tol=tol,
        momentum=momentum,
        nesterovs_momentum=nesterovs_momentum,
        early_stopping=early_stopping,
        validation_fraction=validation_fraction,
        beta_1=beta_1,
        beta_2=beta_2,
        epsilon=epsilon,
        n_iter_no_change=n_iter_no_change,
        max_fun=max_fun,
        random_state=42
    )

    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=99)

    score = cross_val_score(model, X, Y, scoring='accuracy', cv=cv, n_jobs=-1).mean()
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)  

print("Najlepsze parametry:", study.best_params)
print("Najlepszy wynik (accuracy):", study.best_value)
```

```python
from pytorch_tabnet.tab_model import TabNetClassifier
from sklearn.model_selection import RepeatedStratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
import optuna

X = df_imputated.iloc[:, 4:11].values
Y = df_imputated.iloc[:, 13].values

scaler = StandardScaler()
X = scaler.fit_transform(X)

label_encoder = LabelEncoder()
Y = label_encoder.fit_transform(Y)

def objective(trial):
    n_d = trial.suggest_int('n_d', 8, 64)  #rozmiar warstwy decyzyjnej
    n_a = trial.suggest_int('n_a', 8, 64)  #rozmiar warstwy uwagowej
    n_steps = trial.suggest_int('n_steps', 3, 10)  #liczba kroków w TabNet
    gamma = trial.suggest_float('gamma', 1.0, 2.0)  # współczynnik relaksacji
    lambda_sparse = trial.suggest_float('lambda_sparse', 1e-6, 1e-3, log=True)
    learning_rate = trial.suggest_float('learning_rate', 0.01, 0.2, log=True) 

    model = TabNetClassifier(
        n_d=n_d, 
        n_a=n_a, 
        n_steps=n_steps, 
        gamma=gamma,
        lambda_sparse=lambda_sparse, 
        optimizer_params=dict(lr=learning_rate), 
        seed=42
    )

    # Walidacja krzyżowa
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=99)
    scores = []

    for train_idx, test_idx in cv.split(X, Y):
        X_train, X_test = X[train_idx], X[test_idx]
        Y_train, Y_test = Y[train_idx], Y[test_idx]

        model.fit(
            X_train, Y_train,
            eval_set=[(X_test, Y_test)],
            eval_name=["valid"],
            eval_metric=["accuracy"],
            patience=20,
            max_epochs=200,
            batch_size=256,
            virtual_batch_size=64,
        )

        Y_pred = model.predict(X_test)
        score = (Y_pred == Y_test).mean()
        scores.append(score)

    return np.mean(scores)

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1)  

print("Najlepsze parametry:", study.best_params)
print("Najlepszy wynik (accuracy):", study.best_value)
```
