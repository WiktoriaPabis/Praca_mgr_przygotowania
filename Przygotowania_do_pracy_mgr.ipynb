```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                                    random_state= 99)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(criterion = "entropy",
                             max_depth = 8,
                             min_samples_split = 10,
                             random_state = 5)
```

```python
clf.fit(X_train, Y_train)
```

```python
clf.feature_importances_ 
# podglÄ…danie wagi cech w nauczonym modelu
```

```python
Y_pred = clf.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, Y_pred)

# |    TN- dobrze zaklasyfikowana negatywna klasa  |    FP- Åºle zaklasyfikowana klasa pozytywna    |
# |    FN- Åºle zaklasyfikowana klasa negatywna     |    TP- dobrze zaklasyfikowana pozytywna klasa |
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```
